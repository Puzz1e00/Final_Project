```python
days = [1,2,3,4]
max_t = [50,51,52,48]
min_t = [7,12,15,20]
avg = [35,30,45,33]
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather")
plt.plot(days, max_t, label='Max')
plt.plot(days, avg, label= 'Min')
plt.plot(days, avg, label= 'Avg')
plt.legend(loc = "best")
plt.grid()
```
